```python
#There are two types of access operator in pandas 1-loc and 2-iloc ,loc is label based and iloc is index based
data.iloc[1]
```

```python
#in both loc and iloc the first entry is for row and second for column , in both its easy to retrive row as compared to columns
data.iloc[1,2] #first row second column 
```

```python
#select all columns 
data.iloc[:,1] #all rows and first column
```

```python
data.iloc[:3,2] #first three rows and second column 
```

```python
#another way of getting above is 
data.iloc[[0,1,2],2]
```

```python
#getting the resukt from bottom 
data.iloc[-5,1]
```

```python
data.iloc[-4,:] #4th row from bottom and all column 
```

```python
#in all the above example we have dealt with index based selection and now we are going to access the data based upon label based indexing 
data.loc[:,'Region']
```

```python
data.loc[:,1] #if we give index instead of label in loc then it will throw key error as there is no attribute named 1 in the data 
```

```python
data.loc[[1,2,3,4],
         'Region']
```

```python
#getting more than one column 
data.loc[:,['Date','Region']]
```

difference in loc and iloc 
in iloc indexing the first index is included and last one is exculded for example data.iloc[0:10] it will give 0 to 9th index value and in loc both index are inclusive it will give values for index 0 to 10 ,total of 11 values .
