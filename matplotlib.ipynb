```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Basic Graph With Title


```python
x=[1,2,3,4,5,6]
y=[12,13,14,10,18,20]

plt.plot(x,y)
plt.title("Basic Example")
plt.show()
```

Line Plot With Labels

```python
experience = [1,3,4,5,7,8,10,12]
salary = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]
plt.plot(experience,salary)
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

Plotting Multiple Graphs in One Graph

```python
experience = [1,3,4,5,7,8,10,12]
data_scientists_salary = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]
software_engineers_salary = [9020, 12873, 15725, 18000, 19790, 20196, 25769,32000 ]
plt.plot(experience,data_scientists_salary)
plt.plot(experience,software_engineers_salary)
plt.legend(['data_scientists','software_engineers'])
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```

Another way of creating above Graph

```python
experience = [1,3,4,5,7,8,10,12]
data_scientists_salary = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]
software_engineers_salary = [9020, 12873, 15725, 18000, 19790, 20196, 25769,32000 ]
plt.plot(experience,data_scientists_salary, label="data_scientists")
plt.plot(experience,software_engineers_salary, label="software_engineers")
plt.legend(loc="lower right") #  loc is used to locate labels on graph
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```

Colors in Graph

```python
experience = [1,3,4,5,7,8,10,12]
data_scientists_salary = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]
software_engineers_salary = [9020, 12873, 15725, 18000, 19790, 20196, 25769,32000 ]
plt.plot(experience,data_scientists_salary, label="data_scientists", color="y")
plt.plot(experience,software_engineers_salary, label="software_engineers", color="r")
plt.legend(loc="lower right") #  loc is used to locate labels on graph
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```

Adding LineStyle and Marker and LineWidth

```python
experience = [1,3,4,5,7,8,10,12]
data_scientists_salary = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]
software_engineers_salary = [9020, 12873, 15725, 18000, 19790, 20196, 25769,32000 ]
plt.plot(experience,data_scientists_salary, label="data_scientists", color="b", linestyle="--", marker="o", linewidth=1.6)
plt.plot(experience,software_engineers_salary, label="software_engineers", color="r",linestyle=":", marker="o", linewidth=2.6)
plt.legend(loc="lower right") #  loc is used to locate labels on graph
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.tight_layout()
plt.grid(True)
plt.show()
```
